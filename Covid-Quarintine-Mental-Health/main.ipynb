```python
from pandas import *
from matplotlib.pyplot import *
from seaborn import *
```

```python
QM=read_csv("C19-Quarintine-Mental Dataset.csv")
QM
```

```python
'''
EDUCATION: Maximum level of education attained by the individual: 
'Completed postgraduate': 30                
'Incomplete postgraduate': 40               
'Completed tertiary or university': 50      
'Incomplete tertiary or university': 60     
'Completed high school': 70                 
'Incomplete high school': 80                
'Completed elementary school': 90           
'Incomplete elementary school': 100         
'''

def edu(a):
    if a==30:
        return "POSTGRADUATE"
    elif a==40:
        return "NOT POSTGRADUATE"
    elif a==50:
        return "UNIVERSITY"
    elif a==60:
        return "NOT UNIVERSITY"
    elif a==70:
        return "HIGH-SCHOOLER"
    elif a==80:
        return "NOT HIGH-SCHOOLER"
    elif a==90:
        return "ELEMENTARY"
    elif a==100:
        return "NOT ELEMENTARY"
    else:
        return a
    
    
```

```python
QM["EDUCATION"]=QM["EDUCATION"].apply(edu)
QM
```

```python
'''
MENTAL DISORDER HISTORY: 
'no': 0
'yes': 50
'''
def MDH(a):
    if a==50:
        return "YES"
    elif a==0:
        return "NO"
    else:
        return a
```

```python
QM["MENTAL DISORDER HISTORY"]=QM["MENTAL DISORDER HISTORY"].apply(MDH)
QM
```

```python
'''
SUIC ATTEMPT HISTORY:
'ideation': 50
'no': 0
'yes': 100
'''

def SAH(a):
    if a==50:
        return "IDEATION"
    elif a==0:
        return "NO"
    elif a==100:
        return "YES"
    else:
        return a
```

```python
QM["SUIC ATTEMPT HISTORY"]=QM["SUIC ATTEMPT HISTORY"].apply(SAH)
QM
```

```python
'''
LIVING WITH SOMEBODY:
'no': 20
'yes': 0
'''

def LWS(a):
    if a==0:
        return "YES"
    elif a==20:
        return "NO"
    else:
        return a
```

```python
QM["LIVING WITH SOMEBODY"]=QM["LIVING WITH SOMEBODY"].apply(LWS)
QM
```

```python
'''
ECONOMIC INCOME:
'yes': 0
'no': 50.
'''
def EI(a):
    if a==0:
        return "YES"
    elif a==50:
        return "NO"
    else:
        return a
```

```python
QM["ECONOMIC INCOME"]=QM["ECONOMIC INCOME"].apply(EI)
QM
```

```python
QM=QM.sort_values("AGE")
QM
```

```python
lineplot(x="EDUCATION", y="AGE", hue="ECONOMIC INCOME", data=QM)
xticks(rotation=65)
title("Lineplot")
```

```python
histplot(QM['ANXIETY STATE'], kde=True, bins=20)
title("Histplot")
```

```python
figure(figsize=(6,4))
kdeplot(data=QM, x='SUIC RISK', bw_adjust=0.5, fill=True, color='purple', linestyle='--', alpha=0.5)
title("Kdeplot")
```

```python
QM.plot.hexbin(x='ANXIETY STATE', y='AGE', gridsize=10, cmap='viridis')
title("Hexbin")
```

```python
figure(figsize=(7,5))
barplot(x="REGION", y="ANXIETY TRAIT", hue='ECONOMIC INCOME', data=QM)
xticks(rotation=90)
title("Barplot")
```

```python
scatterplot(x="REGION", y="DEPRESSION", hue="ECONOMIC INCOME", data=QM)
xticks(rotation=65)
title("Scatterplot")
```
