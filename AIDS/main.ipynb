```python
from pandas import *
from matplotlib.pyplot import *
from seaborn import *
```

```python
ad=read_csv("AIDS_deaths_state.csv")
ad.head(10)
```

```python
ad["Race"].unique().tolist()
```

```python
ad=ad.drop(["Indicator", "FIPS", "Race", "Sex", "Age group", "Misc"], axis=1)
ad.head(11)
```

```python
# ad["Cases"]=to_numeric(ad["Cases"], errors="coerce")
ad['Cases'] = ad['Cases'].str.replace(',', '').astype(int)
ad['Population'] = ad['Population'].str.replace(',', '').astype(int)
```

```python
ad.info()
```

```python
ad
```

```python
ad["Geography"].unique().tolist()
```

```python
figure(figsize=(10,6))
lineplot(x="Population", y="Cases", color="#FA0A72", data=ad)
title("Line plot")
```

```python
figure(figsize=(15,6))
barplot(x="Geography", y="Population", color="#060270", data=ad)
title("Bar plot")
xticks(rotation=85)
```

```python
kdeplot(data=ad, x='Cases', bw_adjust=0.5, fill=True, color='#178E92', linestyle='--', alpha=0.5)
title("Kdeplot")
```

```python
stripplot(x='Year', y='Cases', data=ad)
xticks(rotation=70)
title("Stripplot")
```

```python
histplot(ad['Population'], kde=True, bins=15)
title("Histplot")
```
